# 数据清洗

## 缺失值处理

```python

df[df['normalized-losses'].isnull()].head()
```

```python
cdf = ECDF(df['normalized-losses'])
plt.subplot(121)
plt.plot(cdf.x,cdf.y)
plt.xlabel('normalized-losses')
plt.ylabel('ECDF')
plt.subplot(122)
plt.hist(df['normalized-losses'])
```

根据累计概率密度以及直方图判断，normalized-losses是非正态分布的，是不能直接使用平均值进行填充的，可以考虑中位数<br>
除此之外，还可以考虑进行分组进行填充，即使用不同组的平均值进行填充。那么要根据什么进行分组呢？可以探究normalized-losses与哪个分类变量相关

```python
df.corr()['normalized-losses'].sort_values(ascending=False)
```

发现normalized-losses与symboling高度相关，可以使用symboling进行分组

```python
df.groupby('symboling')['normalized-losses'].describe()
```

```python
# 使用各组的均值进行填充
df['normalized-losses'] = df.groupby('symboling')['normalized-losses'].transform(lambda x: x.fillna(x.mean()))
```

```python
# 其他属性缺失值较少，可直接删除
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

# 特征工程

## 特征选择

```python
# 数值变量--相关系数
corr = abs(df.corr()['price'])
corr = corr[corr<0.1]
cols_to_drop = corr.index.to_list()
df.drop(cols_to_drop, axis=1,inplace=True)
```

```python
# 分类变量--箱线图
sns.boxplot(y='price',x='fuel-system',data=df)
```

## 特征缩放

```python
df.select_dtypes(include='number').head()
```

```python
features = df.drop(columns='price')
```

```python
num_features = features.select_dtypes(include='number').columns
```

```python
standard_scaler = StandardScaler()
features[num_features] = standard_scaler.fit_transform(features[num_features])
```

## 类别特征转数值类型

```python
categorical_features = features.select_dtypes(include='object')
```

```python
features = features.join(pd.get_dummies(categorical_features))
```

```python
features.drop(columns=categorical_features.columns,inplace=True)
```

```python
features
```

```python
features.columns
```

# 划分训练集、测试集

```python
target = df['price']
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.25,random_state = 42)
```

```python
print("Train", X_train.shape, "and test", X_test.shape)
```

# 模型训练

```python
alphas = 2. ** np.arange(2, 12)
scores = np.empty_like(alphas)
train_scores = np.empty_like(alphas)
for index,alpha in enumerate(alphas):
    lasso = Lasso(random_state=42,alpha=alpha)
    lasso.fit(X_train,y_train)
    scores[index] = lasso.score(X_test, y_test)
    train_scores[index] = lasso.score(X_train,y_train)
lassocv =LassoCV(random_state=42,cv=10)
lassocv.fit(features,target)
lassocv_score = lassocv.score(features, target)
lassocv_alpha = lassocv.alpha_
```

```python
plt.figure(figsize = (10, 4))
plt.plot(alphas, train_scores,'b',label='train_score')
plt.plot(alphas, scores,'r',label='test score')

# plt.axhline(lassocv_score)
plt.xlabel('alpha')
plt.ylabel('CV Score')
plt.xscale('log', basex = 2)
sns.despine(offset = 15)
plt.legend()

print('CV results:', lassocv_score, lassocv_alpha)
```

```python
predicted = lassocv.predict(X_test)
print(r2_score(predicted,y_test))
print(mean_squared_error(predicted,y_test))
```

对比模型在训练集和测试集上的表现，发现模型存在过拟合。<br>
考虑多重共线性

## 多重共线性检测

```python
vif_data = pd.DataFrame()
vif_data["feature"] = features.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(features.values, i)
for i in range(len(features.columns))]
print(vif_data)
```

存在VIF大于10的特征

## 移除具有共线性的特征

```python
df.corr()
```

```python
df['volume'] = df.length * df.width*df.height # 将高度相关的特征进行组合
new_df = df.drop(columns=['length','width'])
```

```python
new_df.corr()
```

```python
new_df.drop(columns=['curb-weight', 'city-mpg','wheel-base',], inplace = True) 
```

```python
standard_scaler = StandardScaler()
features = new_df.drop(columns='price')
num_features = features.select_dtypes(include='number').columns
categorical_features = features.select_dtypes(include='object')
features[num_features] = standard_scaler.fit_transform(features[num_features])
features = features.join(pd.get_dummies(categorical_features))
features.drop(columns=categorical_features.columns,inplace=True)
```

```python
target = df['price']
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.25,random_state = 42)
```

```python
alphas = 2. ** np.arange(2, 12)
scores = np.empty_like(alphas)
train_scores = np.empty_like(alphas)
for index,alpha in enumerate(alphas):
    lasso = Lasso(random_state=42,alpha=alpha)
    lasso.fit(X_train,y_train)
    scores[index] = lasso.score(X_test, y_test)
    train_scores[index] = lasso.score(X_train,y_train)
lassocv =LassoCV(random_state=42,cv=10)
lassocv.fit(features,target)
lassocv_score = lassocv.score(features, target)
lassocv_alpha = lassocv.alpha_
```

```python
plt.figure(figsize = (10, 4))
plt.plot(alphas, train_scores,'b',label='train_score')
plt.plot(alphas, scores,'r',label='test score')

# plt.axhline(lassocv_score)
plt.xlabel('alpha')
plt.ylabel('CV Score')
plt.xscale('log', basex = 2)
sns.despine(offset = 15)
plt.legend()

print('CV results:', lassocv_score, lassocv_alpha)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(alphas, train_scores,'b',label='train_score')
plt.plot(alphas, scores,'r',label='test score')

# plt.axhline(lassocv_score)
plt.xlabel('alpha')
plt.ylabel('CV Score')
# plt.xscale('log', basex = 2)
sns.despine(offset = 15)
plt.legend()

print('CV results:', lassocv_score, lassocv_alpha)
```

```python
print(lassocv_score)
```

```python
# lassocv coefficients
coefs = pd.Series(lassocv.coef_, index = features.columns)

# prints out the number of picked/eliminated features
print("Lasso picked " + str(sum(coefs != 0)) + " features and eliminated the other " +  \
      str(sum(coefs == 0)) + " features.")

# takes first and last 10
coefs = pd.concat([coefs.sort_values().head(5), coefs.sort_values().tail(5)])

plt.figure(figsize = (10, 4))
coefs.plot(kind = "barh")
plt.title("Coefficients in the Lasso Model")
plt.show()
```
