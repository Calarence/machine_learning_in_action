æˆ‘ä»¬åªèƒ½è§‚æµ‹åˆ°ä»¥ä¸Š10æ¬¡ç»“æœï¼Œ**æ¯ä¸€æ¬¡çš„æŠ•æ·ç»“æœéƒ½ä¸çŸ¥é“å®ƒç”±ç¡¬å¸Bè¿˜æ˜¯ç¡¬å¸Cäº§ç”Ÿçš„**ï¼Œç°åœ¨æˆ‘ä»¬éœ€è¦æ±‚çš„æ˜¯ä¸‰æšç¡¬å¸å‡ºç°æ­£é¢çš„æ¦‚ç‡ï¼Œä¹Ÿå³æ˜¯ä¸‰æšç¡¬å¸çš„æ¨¡å‹å‚æ•°ã€‚

åœ¨è¿™ä¸ªæ¨¡å‹ä¸­ï¼Œéšå˜é‡å°±æ˜¯ç¡¬å¸Açš„æŠ•æ·ç»“æœï¼Œå®ƒç”¨æˆ·æŒ‡å¯¼æŠ•æ·ç¡¬å¸Bè¿˜æ˜¯ç¡¬å¸Cã€‚å› æ­¤æˆ‘ä»¬æœ‰ï¼š

$$


yä»£è¡¨æœ€åè§‚æµ‹å¹¶è®°å½•ä¸‹æ¥çš„ç»“æœï¼Œå³0æˆ–è€…1ï¼›zæ˜¯éšå˜é‡ï¼Œå³æˆ‘ä»¬æœªè®°å½•ä¸‹æ¥çš„ç¡¬å¸Açš„è§‚æµ‹ç»“æœ(æ˜¯æˆ‘ä»¬ä¸çŸ¥é“ç»“æœ)ï¼›$\theta=(\pi,p,q)$æ˜¯æ•´ä¸ªæ¨¡å‹çš„å‚æ•°ï¼Œå³A,B,Cä¸‰æšç¡¬å¸ä¸ºæ­£é¢çš„æ¦‚ç‡ã€‚


$$
P(1|\theta)=\pi{p}+(1-\pi){q}
$$


$$
P(0|\theta)=\pi{(1-p)}+(1-\pi){(1-q)}
$$


æ ¹æ®æå¤§ä¼¼ç„¶ä¼°è®¡æˆ‘ä»¬å¯ä»¥å¾—åˆ°ï¼š


$$
P(Y|\theta)=\prod \pi{p^y_i}(1-p)^{1-y_i}+(1-\pi)q^y_i(1-q)^{1-y_i}
$$



$Y=(y_1,y_2,...y_n)$ä¸ºæ¯æ¬¡è§‚æµ‹åˆ°çš„ç»“æœã€‚æˆ‘ä»¬éœ€è¦æ±‚è§£çš„æ˜¯åœ¨å‚æ•°$\theta$ä¸ºä»€ä¹ˆçš„æƒ…å†µä¸‹ï¼Œä¸Šå¼çš„å–å€¼æœ€å¤§ã€‚



é—æ†¾çš„æ˜¯ï¼Œä¸Šå¼å¹¶æ²¡æœ‰è§£æè§£ï¼Œæˆ‘ä»¬éœ€è¦é€šè¿‡è¿­ä»£çš„æ–¹æ³•æ‰èƒ½å¾—åˆ°ä¸€ä¸ªæ¯”è¾ƒå¥½çš„è¿‘ä¼¼è§£ã€‚ä¹Ÿå°±æ˜¯æˆ‘ä»¬è¯´çš„EMç®—æ³•ã€‚

#### EMç®—æ³•æ­¥éª¤å¦‚ä¸‹ï¼š

(1):åˆå§‹åŒ–æ¨¡å‹å‚æ•°$\theta$,æ¯”å¦‚æˆ‘ä»¬å¯ä»¥$\theta=(\pi=0.5,p=0.5,q=0.5)$;

(2):Eæ­¥ï¼Œè®°$\theta^{(i)}$ä¸ºç¬¬iæ¬¡è¿­ä»£å‚æ•°$\theta$çš„ä¼°è®¡å€¼ï¼Œåœ¨ç¬¬$i+1$æ¬¡è¿­ä»£çš„Eæ­¥ä¸­ï¼Œè®¡ç®—ï¼š


$$
Q(\theta,\theta^{(i)})=E_Z[logP(Y,Z|\theta)Y,\theta^{(i)}]=\sum logP(Y,Z|\theta)P(Z|Y,\theta^{(i)})
$$


è¿™ä¸ªå¼å­çœ‹èµ·æ¥æœ‰ç‚¹å¤æ‚ï¼Œå…¶å®å°±æ˜¯æ ¹æ®æˆ‘ä»¬ä¸Šä¸€æ­¥ä¼°è®¡çš„å‚æ•°$\theta^{(i)}$ä»¥åŠè§‚æµ‹ç»“æœ Y æ¥ä¼°è®¡éšå˜é‡çš„ç»“æœã€‚

åœ¨æˆ‘ä»¬ä¾‹å­ä¸­å°±æ˜¯æˆ‘ä»¬ç°åœ¨çŸ¥é“äº†æ¯ä¸ªç¡¬å¸çš„å‚æ•°ä»¥åŠæœ€åçš„è§‚æµ‹ç»“æœï¼Œæ¥ä¼°è®¡éšå˜é‡$Z$,


$$
P(z_j|\theta^i,y_j)=\frac{{\pi}^i(p^i)^{y_j} (1-p^i)^{(1-y_j)}}{{\pi}^i(p^i)^{y_j} (1-p^i)^{(1-y_j)}+(1-{\pi}^i)(q^i)^{y_j}(1-q^i)^(1-y_j)}
$$


$P(z_j|\theta^i,y_j)$å°±æ˜¯ç¡®å®šå‚æ•°å’Œè§‚æµ‹ç»“æœyçš„æƒ…å†µä¸‹ï¼Œè§‚æµ‹ç»“æœæ¥è‡ªäºç¡¬å¸Bçš„æ¦‚ç‡ã€‚æˆ‘ä»¬å¯ä»¥å†™ä¸€ä¸ªå‡½æ•°func_Eæ¥è®¡ç®—ã€‚



(3):Mæ­¥,æ±‚ä½¿å¾—$Q(\theta,\theta^i )$æœ€å¤§åŒ–çš„$\theta$ï¼Œç¡®å®šç¬¬$i+1$æ¬¡è¿­ä»£æ—¶å€™çš„å‚æ•°å€¼$\theta^{i+1}$

åœ¨Eæ­¥ä¸­æˆ‘ä»¬æ±‚å‡ºäº†éšå˜é‡çš„ç»“æœï¼Œæ ¹æ®éšå˜é‡çš„ç»“æœæˆ‘ä»¬åŒæ ·å¯ä»¥æ±‚å‡ºå‚æ•°$\theta$çš„å€¼ï¼š


$$
\pi^{i+1}=\frac{1}{n}\sum_{j=1}^{n} \mu_j^{i+1}
$$



$$
p^{i+1}=\frac{\sum_{j=1}^{n} \mu_j^{i+1}y_j}{\sum_{j=1}^{n} \mu_j^{i+1}}
$$



$$
q^{i+1}=\frac{\sum_{j=1}^{n} (1-\mu_j^{i+1})y_j}{\sum_{j=1}^{n} (1-\mu_j^{i+1})}
$$


è¿™ä¸‰ä¸ªå¼å­å¾ˆå¥½ç†è§£ï¼Œæˆ‘ä»¬æŒ‡å¯¼äº†æ¯æ¬¡è§‚æµ‹ç»“æœæ¥è‡ªäºç¡¬å¸Bçš„æ¦‚ç‡ä¹‹åï¼Œå¾ˆè‡ªç„¶åœ°é€šè¿‡æå¤§ä¼¼ç„¶ä¼°è®¡å¯ä»¥å¾—åˆ°$\pi,p,q$ã€‚

åŒæ ·åœ°ï¼Œæˆ‘ä»¬å¯ä»¥å†™ä¸€ä¸ªå‡½æ•°func_Mæ¥è®¡ç®—Mæ­¥çš„å€¼

(4)é‡å¤EMæ­¥ï¼Œç›´åˆ°æ”¶æ•›

```python
def func_E(theta,y):
    pi,p,q=theta
    if y==1:
        u=(pi*p)/(pi*p+(1-pi)*q) 
        #å½“è§‚æµ‹ç»“æœæ˜¯1çš„æ—¶å€™ï¼Œæ¥è‡ªäºç¡¬å¸Bçš„æ¦‚ç‡å°±æ˜¯ä½¿ç”¨ç¡¬å¸BæŠ•æ·å‡º1çš„æ¦‚ç‡(å³pi*p)é™¤ä»¥è§‚æµ‹ç»“æœå‡ºç°1çš„æ¦‚ç‡(pi*p+(1-pi)*q)
    else:
        u=(pi*(1-p))/(pi*(1-p)+(1-pi)*(1-q))
        #å½“è§‚æµ‹ç»“æœæ˜¯0çš„æ—¶å€™ï¼Œæ¥è‡ªäºç¡¬å¸Bçš„æ¦‚ç‡å°±æ˜¯ä½¿ç”¨ç¡¬å¸BæŠ•æ·å‡º0çš„æ¦‚ç‡(å³pi*(1-p))é™¤ä»¥è§‚æµ‹ç»“æœå‡ºç°0çš„æ¦‚ç‡(pi*(1-p)+(1-pi)*(1-q))
    return u
U=[]
Y=[1,1,0,1,0,0,1,0,1,1]
theta=(0.5,0.5,0.5)
for i in Y:   
    u=func_E(theta,i)
    U.append(u)
print(U)
```

```python
def func_M(U,Y):
    pi=sum(U)/len(U)
    p=sum([U[i]*Y[i] for i in range(len(U))])/sum(U)
    q=sum([(1-U[i])*Y[i] for i in range(len(U))])/sum([1-i for i in U])
    return (pi,p,q)
U=[0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]
Y=[1,1,0,1,0,0,1,0,1,1]
theta=func_M(U,Y)
print(theta)
```

å°†ä¸¤ä¸ªå‡½æ•°ç»¼åˆèµ·æ¥ï¼Œå¯¹äºè¯¥é—®é¢˜ï¼Œæ•´ä¸ªæ¨¡å‹çš„ä»£ç ä¸ºï¼š

```python
class EM:
    def __init__(self,theta,Y):
        self.theta=theta
        self.Y=Y
        self.u=None
    def func_E(self,theta):
        pi,p,q=theta
        U=[]
        for i in self.Y:
            if i==1:
                u=(pi*p)/(pi*p+(1-pi)*q) 
            else:
                u=(pi*(1-p))/(pi*(1-p)+(1-pi)*(1-q))
            U.append(u)
        return U
    
    def func_M(self,U):
        pi=sum(U)/len(U)
        p=sum([U[i]*self.Y[i] for i in range(len(U))])/sum(U)
        q=sum([(1-U[i])*self.Y[i] for i in range(len(U))])/sum([1-i for i in U])  
        
        return (pi,p,q)
    
    def train(self):
        for i in range(10):
            U=self.func_E(self.theta)
            theta=self.func_M(U)
            print("ç¬¬{}æ¬¡è¿­ä»£çš„ç»“æœï¼š".format(i+1),theta)
        
theta=(0.4,0.6,0.7)
Y=[1,1,0,1,0,0,1,0,1,1]
model1=EM(theta,Y)
model1.train()
theta=(0.4,0.6,0.3)
model2=EM(theta,Y)
model2.train()
```

## ä¸‰ã€ä½¿ç”¨EMç®—æ³•æ¥æ±‚è§£K-means

ä»‹ç»å®Œäº†EMç®—æ³•ä¹‹åï¼Œæˆ‘ä»¬å°±éœ€è¦ä½¿ç”¨EMç®—æ³•æ¥æ±‚è§£k-meansç®—æ³•ã€‚åœ¨æ­¤ä¹‹å‰æˆ‘ä»¬å…ˆé‡‡ç”¨EMç®—æ³•æ¥è§£é‡ŠK-meansç®—æ³•ã€‚

åœ¨k-meansç®—æ³•ä¸­æˆ‘ä»¬å¯ä»¥å°†æ¯ä¸ªé›†åˆä¸­å¿ƒç‚¹è§†ä¸ºä¸ºéšå˜é‡ï¼Œæ¯ä¸ªæ ·æœ¬çš„ç±»åˆ«æ˜¯æˆ‘ä»¬æœ€åæ‰€æ±‚çš„ï¼Œå¯ä»¥è§†ä¸ºå‚æ•°ã€‚

æ³¨ï¼šåœ¨å®é™…çš„è®¡ç®—ä¸­ï¼Œæˆ‘ä»¬å°†ä¸¤æ­¥é¢ å€’ä¹Ÿæ˜¯å¯ä»¥çš„ï¼Œé¦–å…ˆåˆå§‹åŒ–ä¸­å¿ƒç‚¹ã€‚

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
X=datasets.load_iris()['data'][:,:2]
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
ax.scatter(X[:,0],X[:,1])
plt.show()
```

ä»¥ä¸Šå°±æ˜¯æˆ‘ä»¬çš„æ•°æ®ï¼Œæˆ‘ä»¬éœ€è¦ä½¿ç”¨EMç®—æ³•å°†å…¶åˆ†ä¸º3ç±»ã€‚ä»£ç å¦‚ä¸‹æ‰€ç¤ºï¼š

```python
class KMeans:
    def __init__(self,K,X):
        """
        K:ç±»åˆ«çš„æ•°é‡
        X:æ•°æ®
        """
        self.K=K
        self.X=X
        self.w=X.shape[1] ##æ•°æ®Xçš„ç»´åº¦
        print(self.K,X.shape[1])
        self.center=np.zeros([self.K,X.shape[1]]) ##ä¸­å¿ƒç‚¹
        self.cate=None                      ##æ¯æ¡æ•°æ®çš„ç±»åˆ«
        
    def cal_eclud(self,vec1,vec2):
        return sum((vec1-vec2)**2)**0.5
        
        
    def initialization(self):
        #ä¸­å¿ƒç‚¹çš„åˆå§‹åŒ–
        self.center=self.X[:self.K] #éšæœºå°†Xä¸­å‰Kä¸ªå€¼å¤åˆ¶ç»™ä¸­å¿ƒç‚¹ã€‚
        
    def update_cate(self):
        #æ ¹æ®ä¸Šä¸€æ­¥æ±‚å¾—çš„ä¸­å¿ƒç‚¹ï¼Œæ›´æ–°æ¯ä¸ªæ ·æœ¬çš„ç±»åˆ«
        dis=np.zeros([self.X.shape[0],self.K])
        for i in range(self.X.shape[0]):
            for j in range(self.K):
                dis[i,j]=self.cal_eclud(self.X[i],self.center[j])
        self.cate=np.argmin(dis,axis=1)
    
    def update_center(self):
        #æ ¹æ®æ¯ä¸ªæ ·æœ¬çš„ç±»åˆ«æ¥é‡æ–°æ›´æ–°ä¸­å¿ƒç‚¹
        for i in range(self.K):
            self.center[i]=np.mean(self.X[self.cate==i],axis=0)
    def train(self):
        self.initialization()
        for i in range(15):
            self.update_cate()
            self.update_center()
            print("ç¬¬ä¸€ä¸ªä¸­å¿ƒç‚¹çš„ä½ç½®æ˜¯ï¼š",self.center[0])
        
model=KMeans(3,X)     
model.train()  
cate=model.cate
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
id_1=np.where(cate==0)
p1=ax.scatter(X[id_1,0],X[id_1,1],marker=".",color="green",s=180)
id_2=np.where(cate==1)
p2=ax.scatter(X[id_2,0],X[id_2,1],marker=".",color="black",s=180)
id_3=np.where(cate==2)
p3=ax.scatter(X[id_3,0],X[id_3,1],marker=".",color="red",s=180)
plt.show()
```
